```python
# We detected the outliers.
```

```python


for column in df.columns:
    if pd.to_numeric(df[column], errors='coerce').notnull().all():
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        outliers = df[(df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))]

        if not outliers.empty:
            print(f"Outliers found! Column: {column}")
            print(outliers)
            print('\n')
        else:
            print(f"No outliers found. Column: {column}\n")
    else:
        print(f"{column} column does not contain numerical values. Skipped.\n")
```
