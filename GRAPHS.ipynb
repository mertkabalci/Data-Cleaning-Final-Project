```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python


df = pd.read_excel(r"C:\Users\GHOST\Downloads\Cleaned_Data.xlsx")


size_factor = 500  


plt.figure(figsize=(10, 6))  


region_markers = {'North': 'o', 'South': 's', 'East': 'D', 'West': '^'}


for region, marker in region_markers.items():
    region_data = df[df['Region'] == region]
    sns.scatterplot(x='Customer_Count', y='Units_Sold', size='Customer_Count', sizes=(50, 200),
                    marker=marker, label=region, data=region_data, alpha=1)


plt.xlabel('Customer Count')
plt.ylabel('Units Sold')
plt.title('Bubble Plot of Units Sold vs Customer Count by Region')


legend = plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
legend.get_title().set_fontsize('10')  


plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d.%m.%Y')

df['Year'] = df['Date'].dt.year


grouped_data = df.groupby('Year')['Revenue'].sum().reset_index()


plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Revenue', data=grouped_data)
plt.title('Total Revenue Over the Years')
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.show()

```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.barplot(x="Region", y="Revenue", hue="Product_Category", data=df)

plt.title('Relationship between region, product category and revenue')
plt.show()
```

```python

plt.figure(figsize=(10, 6)) 

sns.boxplot(x='Region', y='Units_Sold', data=df, showfliers=False)

plt.title('Boxplot of Units Sold by Region')

plt.show()
```

```python
df['Combined_Score'] = (df['Employee_Rating'] + df['Customer_Satisfaction']) / 2
correlation_coefficient_combined = df['Combined_Score'].corr(df['Units_Sold'])
plt.figure(figsize=(8, 6))
sns.regplot(x='Combined_Score', y='Units_Sold', data=df, scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.title('Scatter Plot and Regression Line\nCorrelation Coefficient: {:.2f}'.format(correlation_coefficient_combined))
plt.xlabel('Combined Score (Employee Rating + Customer Satisfaction)')
plt.ylabel('Units Sold')
plt.show()
```

```python
correlation_coefficient = df['Employee_Rating'].corr(df['Customer_Satisfaction'])
plt.figure(figsize=(8, 6))
sns.regplot(x='Employee_Rating', y='Customer_Satisfaction', data=df, scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.title('Scatter Plot and Regression Line\nCorrelation Coefficient: {:.2f}'.format(correlation_coefficient))
plt.xlabel('Employee Rating')
plt.ylabel('Customer Satisfaction')
plt.show()
```
